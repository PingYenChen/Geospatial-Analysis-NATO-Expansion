## Part 1: Plot the Current NATO Members

```python
#Read the file, file can be found "https://www.naturalearthdata.com/downloads/"
world = gpd.read_file("ne_110m_admin_0_countries.zip")
world = world[['SOVEREIGNT', 'CONTINENT', 'GDP_MD', 'geometry', 'ISO_A3', 'POP_EST']].copy()

# Rename the columns
world = world.rename(columns={
    'SOVEREIGNT': 'name',
    'CONTINENT': 'continent',
    'GDP_MD': 'gdp_md_est',
    'ISO_A3': 'iso_a3',
    'POP_EST': 'pop_est'
})
```

## Part 2: Potential New Joins Set 1 and Set 2 Analysis

```python
set1 = ['Austria','Cyprus','Kosovo','Ireland','Malta'] #Potential new joins, more likely to happen
```

```python
set2 =['Republic of Serbia','Ukraine','Moldova','Switzerland','Georgia']  #Also potential new joins, but less likely to happen
```

```python
world.name
```

```python
world.name.isin(set1)
```

```python
yellow_countries = world[world.name.isin(set1)]
```

```python
yellow_countries
```

```python
world.name.isin(set2)
```

```python
grey_countries=world[world.name.isin(set2)]
grey_countries
```

```python
# Plot the potential new joins set 1
fig, ax =plt.subplots(figsize =(18,10))
yellow_countries.plot(color='yellow', ec ='black', ax = ax, label ='Set 1')
```

```python
# Plot current members with Set 1 
fig, ax = plt.subplots(figsize =(18,10))
world.plot(ax = ax, color ='whitesmoke', ec = 'black')
world_nato.plot(ax = ax, color = "blue", ec ='black', label ='NATO')
yellow_countries.plot(color='yellow', ec ='black', ax = ax, label ='Set 1')
```

```python
#Potential new join Set 2
fig, ax =plt.subplots(figsize =(18,10))
grey_countries.plot(color='grey', ec ='black', ax = ax, label ='Set 2')
```

```python
#Plot current members with potential Set 1 and Set 2 new joins, we can then see the NATO members before and after
fig, ax = plt.subplots(figsize =(20,10))
world.plot(ax = ax, color ='whitesmoke', ec = 'black', label = "World Map")
world_nato.plot(ax = ax, color = "blue", ec ='black', label ='NATO')
yellow_countries.plot(color='yellow', ec ='black', ax = ax, label ='Set 1')
grey_countries.plot(color='grey', ec ='black', ax = ax, label ='Set 2')
```

### Add legend to the map

```python
ax
```

```python
ax.collections
```

```python
for t in ax.collections:
    print("\n\nt= ",t, ",nt.get_facecolor()=", t.get_facecolor(), "\nt.getlabel=", t.get_label())
```

```python
from matplotlib.lines import Line2D
```

```python
lines = [ Line2D(
                 [0],
                 [0],
                 linestyle ='None',
                 marker ='s',
                 markerfacecolor = t.get_facecolor())
    for t in ax.collections[1:]
        ]
```

```python
lines
```

```python
labels = [t.get_label() for t in ax.collections[1:]]
labels
```

```python
fig, ax = plt.subplots(figsize =(20,10))
world.plot(ax = ax, color ='whitesmoke', ec = 'black', label = "World Map")
world_nato.plot(ax = ax, color = "blue", ec ='black', label ='NATO')
yellow_countries.plot(color='yellow', ec ='black', ax = ax, label ='Set 1')
grey_countries.plot(color='grey', ec ='black', ax = ax, label ='Set 2')

ax.legend(lines,
         labels,
         bbox_to_anchor =(1.15,1))
plt.show()
```
